```python
import sympy as sym
import numpy as np
import time
import cloudpickle

m = [0]*3

with open("Feasible_Solution/0/steady-state.pkl", "rb") as f:
    m[1] = cloudpickle.load(f)
    
with open("Feasible_Solution/0/accel.pkl", "rb") as f:
    m[0] = cloudpickle.load(f)
    
with open("Feasible_Solution/0/decel.pkl", "rb") as f:
    m[2] = cloudpickle.load(f)
    
N = m[0].N[-1]

cN = m[0].cN[-1]

adder = 0
body_xb = []
body_yb = []
body_xf = []
body_yf = []
leg_xt1_L = []
leg_yt1_L = []
leg_xb1_L = []
leg_yb1_L = []
leg_xt2_L = []
leg_yt2_L = []
leg_xb2_L = []
leg_yb2_L = []
leg_xt1_R = []
leg_yt1_R = []
leg_xb1_R = []
leg_yb1_R = []
leg_xt2_R = []
leg_yt2_R = []
leg_xb2_R = []
leg_yb2_R = []

values = []
positions = []

time0 = []
adder0 = 0
val = 0
for i in range(0, 3):
    if(i>0):
        adder0 += (m[i-1].tt0[N].value - m[i-1].tt0[1].value)
    for n in range(1, N+1):
        val = m[i].tt0[n].value - m[i].tt0[1].value
        time0.append(val + adder0)
        
for i in range(0, 3):
    if(i>0):
        adder += m[i-1].q0[N,'x'].value
    for n in range(1, N+1):
        for c in range(1, cN):
            body_xb.append(m[i].q[n,c,'x'].value - 0.5*m[i].lb*np.cos(0) + adder)
            body_yb.append(m[i].q[n,c,'z'].value - 0.5*m[i].lb*np.sin(0))
            body_xf.append(m[i].q[n,c,'x'].value + 0.5*m[i].lb*np.cos(0) + adder)
            body_yf.append(m[i].q[n,c,'z'].value + 0.5*m[i].lb*np.sin(0))  

            # left-------------------------------------------------------------------------------------------------- 
            leg_xt1_L.append(m[i].q[n,c,'x'].value + adder)
            leg_yt1_L.append(m[i].q[n,c,'z'].value) 
            leg_xb1_L.append(m[i].q[n,c,'x'].value + m[i].ll1*np.cos(m[i].q[n,c,'theta_l_L'].value) + adder)
            leg_yb1_L.append(m[i].q[n,c,'z'].value - m[i].ll1*np.sin(m[i].q[n,c,'theta_l_L'].value))

            leg_xt2_L.append(m[i].q[n,c,'x'].value + m[i].ll2*np.cos(m[i].q[n,c,'theta_l_L'].value) + adder)
            leg_yt2_L.append(m[i].q[n,c,'z'].value - m[i].ll2*np.sin(m[i].q[n,c,'theta_l_L'].value))
            leg_xb2_L.append(m[i].q[n,c,'x'].value + (m[i].ll2 + m[i].q[n,c,'r_L'].value)*np.cos(m[i].q[n,c,'theta_l_L'].value) + adder)
            leg_yb2_L.append(m[i].q[n,c,'z'].value - (m[i].ll2 + m[i].q[n,c,'r_L'].value)*np.sin(m[i].q[n,c,'theta_l_L'].value))

            # right---------------------------------------------------------------------------------------------------
            leg_xt1_R.append(m[i].q[n,c,'x'].value + adder)
            leg_yt1_R.append(m[i].q[n,c,'z'].value) 
            leg_xb1_R.append(m[i].q[n,c,'x'].value + m[i].ll1*np.cos(m[i].q[n,c,'theta_l_R'].value) + adder)
            leg_yb1_R.append(m[i].q[n,c,'z'].value - m[i].ll1*np.sin(m[i].q[n,c,'theta_l_R'].value)) 

            leg_xt2_R.append(m[i].q[n,c,'x'].value + m[i].ll2*np.cos(m[i].q[n,c,'theta_l_R'].value) + adder)
            leg_yt2_R.append(m[i].q[n,c,'z'].value - m[i].ll2*np.sin(m[i].q[n,c,'theta_l_R'].value))
            leg_xb2_R.append(m[i].q[n,c,'x'].value + (m[i].ll2 + m[i].q[n,c,'r_R'].value)*np.cos(m[i].q[n,c,'theta_l_R'].value) + adder)
            leg_yb2_R.append(m[i].q[n,c,'z'].value - (m[i].ll2 + m[i].q[n,c,'r_R'].value)*np.sin(m[i].q[n,c,'theta_l_R'].value))
            
time = []
adder = 0
for i in range(0, 3):
    if(i>0):
        adder += (m[i-1].tt[N,cN].value - m[i-1].tt[1,1].value)
    for n in range(1, N+1):
        for c in range(1, cN):
            time.append(m[i].tt[n,c].value - m[i].tt[1,1].value + adder)
            
def time_interval():
    max_val = 0
    for a in range(0, len(time)-1):
        this = time[a+1] - time[a]
        if(abs(this)>max_val):
            max_val = abs(this)
    return abs(max_val)

# print(time_interval())

max_val = 0
for i in range(0,3):
    for n in range(1, N+1):
        max_val += m[i].h[n].value
max_val = max_val/N*1000
print(time[-1]*1000/len(body_xb))
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as ani
from IPython.display import HTML

# create a time array from 0..t_stop sampled at 0.02 second steps
dt = 0.02

fig = plt.figure() #figsize=(5, 4)
ax = fig.add_subplot(autoscale_on=False, xlim=(-0.5, 6), ylim=(0, 1.))
# ax.set_aspect('equal')
ax.grid()

line_B, = ax.plot([], [], '-', lw=2)
line_L1, = ax.plot([], [], '-', lw=2)
line_L2, = ax.plot([], [], '-', lw=2)
line_R1, = ax.plot([], [], '-', lw=2)
line_R2, = ax.plot([], [], '-', lw=2)

time_template = 'time = %.1fs'
time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)

def animate(i):
    thisx = [body_xb[i], body_xf[i]]
    thisy = [body_yb[i], body_yf[i]]
    line_B.set_data(thisx, thisy)
    
    thisx = [leg_xt1_L[i], leg_xb1_L[i]]
    thisy = [leg_yt1_L[i], leg_yb1_L[i]]
    line_L1.set_data(thisx, thisy)  
    
    thisx = [leg_xt2_L[i], leg_xb2_L[i]]
    thisy = [leg_yt2_L[i], leg_yb2_L[i]]
    line_L2.set_data(thisx, thisy)
    
    thisx = [leg_xt1_R[i], leg_xb1_R[i]]
    thisy = [leg_yt1_R[i], leg_yb1_R[i]]
    line_R1.set_data(thisx, thisy)
    
    thisx = [leg_xt2_R[i], leg_xb2_R[i]]
    thisy = [leg_yt2_R[i], leg_yb2_R[i]]
    line_R2.set_data(thisx, thisy)    
    
    time_text.set_text(time_template % (time[i]))
    return line_B, line_L1, line_L2, line_R1, line_R2, time_text

animate = ani.FuncAnimation(fig, animate, len(body_xb), interval=time[-1]*1000/len(body_xb), blit=True, repeat=False)
HTML(animate.to_jshtml())
# HTML(animate.to_html5_video())
```
